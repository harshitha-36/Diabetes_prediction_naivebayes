```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/boat.jpg")
```

```python
#Get the image type
print("Image type:", img.dtype)

# Get the image resolution (width, height)
image_resolution = img.shape[:2]
print("Image resolution:", image_resolution)
```

```python
plt.imshow(img)
```

```python
#converting image from bgr to rgb
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("BGR")
plt.subplot(1,2,2)
plt.imshow(img1)
plt.title("RGB")
```

```python
planes = cv2.split(img)

# Find the min and max values of each plane
min_values = [plane.min() for plane in planes]
max_values = [plane.max() for plane in planes]

print("Minimum values:", min_values)
print("Maximum values:", max_values)
```

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("BGR")
plt.subplot(1,2,2)
plt.imshow(gray,cmap='gray')
plt.title("GRAY")
```

```python
plt.imshow(img1)
plt.title("COLOR")
```

```python
#Image Crop
img2=img1[200:375,300:450]
plt.imshow(img2)
plt.title("CROP")
```

```python
#Increasing Brightness of a image
img3=img + 20
```

```python
plt.subplot(2,2,1)
plt.imshow(img)
plt.title("Before")
plt.subplot(2,2,2)
plt.imshow(img3)
plt.title("After")
```
